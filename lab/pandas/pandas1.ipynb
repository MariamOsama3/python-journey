```python
csvfile2=pd.read_csv("D:/LAB01/LAB01/needed_files/WA_Fn-UseC_-Telco-Customer-Churn.csv", index_col="customerID")
csvfile2.head()
```

```python
csvfile2['TotalCharges'] = pd.to_numeric(csvfile2['TotalCharges'], errors='coerce')
csvfile2.info()
print("\n")
csvfile2['TotalCharges'].isnull().any()
```

```python
pd.get_dummies(csvfile2['TotalCharges'])
```

```python
mean1=csvfile2['TotalCharges'].mean()
csvfile2['TotalCharges']=csvfile2['TotalCharges'].fillna(mean1)
csvfile2['TotalCharges'].isnull().any()
```

```python
csvfile2.info()
```

```python
newcolumns=pd.get_dummies(csvfile2[['gender','Partner','Dependents','PhoneService','MultipleLines','InternetService',
                        'OnlineSecurity','DeviceProtection','TechSupport','StreamingTV','StreamingMovies',
                        'PaperlessBilling','PaymentMethod']])
```

```python
csvfile2 = csvfile2.drop(['gender','Partner','Dependents','PhoneService','MultipleLines','InternetService',
                        'OnlineSecurity','DeviceProtection','TechSupport','StreamingTV','StreamingMovies',
                        'PaperlessBilling','PaymentMethod'], axis=1)
csvfile2.head()
```

```python
csvfile2['Contract'].value_counts()
```

```python
encoding={'Month-to-month':1,'One year':2,'Two year':3}
csvfile2['Contract']=csvfile2['Contract'].replace(encoding)
csvfile2['Contract']
```

```python
csvfile2['Contract']=pd.to_numeric(csvfile2['Contract'], errors='coerce')
csvfile2.info()
print("\n")
csvfile2['Contract'].isnull().any()
```

```python
encoding={'Yes':1,'No':0}
csvfile2['Churn']=csvfile2['Churn'].replace(encoding)
csvfile2['Churn']
```

```python
csvfile2['Churn']=pd.to_numeric(csvfile2['Churn'], errors='coerce')
csvfile2.info()
print("\n")
csvfile2['Churn'].isnull().any()
```
